# Revisar Correlación e Importancia de Variables

Se transforma las variables categóricas en dummies para trabajarlas 

```python
cats = df.dtypes.to_frame().reset_index()
cats.columns = ['cols','type']
cat_cols = cats[cats.type == 'object'].cols.tolist()

df = pd.get_dummies(df, columns=cat_cols)
df = df.astype(float)
```

Utilización del modelo que se trabajará en este código para obtener la importancia de cada feature

```python
X = df.drop('target', axis=1)
y = df['target']
# Entrenar un modelo de ejemplo
# model = RandomForestClassifier()
model = BalancedRandomForestClassifier(random_state= seed,replacement=False,bootstrap=True,sampling_strategy='all')
model.fit(X, y)

# Obtener la importancia de las características
importances = model.feature_importances_

# Crear un DataFrame para visualización
feature_importances = pd.DataFrame({'Feature': X.columns, 'Importance': importances})
feature_importances = feature_importances.sort_values(by='Importance', ascending=False)

sns.barplot(x='Importance', y='Feature', data=feature_importances)
plt.show()
```

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm',)
```

Se decide dropear la variable "x2" dada su alta correlación con "x1", siendo esta última de mayor importancia que "x2"

```python
df1 = df.drop('x2',axis=1)
```

Normalización de todas las variables que no son dummies

```python
scaler = StandardScaler()
# df1[df1.drop(['target','x1'],axis=1).columns] = scaler.fit_transform(df1.drop(['target','x1'],axis=1))
df1[['x1','x5','x6','x7','x8']] = scaler.fit_transform(df1[['x1','x5','x6','x7','x8']])
# df1
```

Ranking de Variables Dummies con Chi Cuadrado

```python
X_encoded = X[['x3_QKP', 'x3_SAT', 'x3_XJB','x4_MZBER', 'x4_PQKE', 'x4_YEQA', 'x4_ZUQF']]
chi2_selector = SelectKBest(chi2, k='all')  # k='all' para obtener puntajes para todas las características
chi2_selector.fit(X_encoded, y)

# Mostrar puntajes de Chi-Cuadrado
chi2_scores = pd.DataFrame({
    'Feature': X_encoded.columns,
    'Chi2 Score': chi2_selector.scores_
}).sort_values(by='Chi2 Score', ascending=False)

print(chi2_scores)
```

Ranking de Variables con RFE

```python

brf = BalancedRandomForestClassifier(random_state= seed,replacement=False,bootstrap=True,sampling_strategy='all')
rfe = RFE(estimator=brf, n_features_to_select=1)
# Ajustar RFE con nuestros datos
rfe.fit(X, y)
# Filtrar las características seleccionadas
sel = pd.DataFrame(data=[X.columns.tolist(),rfe.ranking_]).T.sort_values(1)
sel.columns = ['Columnas','Ranking']
sel
```

Finalmente de acuerdo a SelectKBest y los métodos anteriormente utlizados, se decide ocupar las mejores 8 features mostradas en la siguiente celda

```python
X = df1.drop('target', axis=1)
y = df1['target']
column_names = X.columns.tolist()  # Lista de nombres de columnas

# Aplicar SelectKBest para seleccionar las mejores características
selector = SelectKBest(score_func=f_classif, k=8)
selector.fit(X, y)

# Obtener los nombres de las columnas seleccionadas usando los índices de get_support()
selected_features = [column_names[i] for i in range(len(column_names)) if selector.get_support()[i]]

print("Columnas seleccionadas:")
print(selected_features)
```

Se genera el dataset con las variables seleccionadas

```python
selector = SelectKBest(score_func=f_classif, k=8)
X_new = selector.fit_transform(X, y)

X = X_new
y = df1['target']
```

# Búsqueda de los mejores hiperparámetros

```python
%%time
smote_enn = SMOTEENN(random_state=seed)

param_grid = {
    'n_estimators': [280],
    'max_depth': [5,6,7],
    'min_samples_split': [2]
}

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y,random_state= seed)

# X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.15,stratify=y_train, random_state=seed)
X_resampled, y_resampled = smote_enn.fit_resample(X_train, y_train)

brf = BalancedRandomForestClassifier(random_state= seed ,replacement=False,bootstrap=True,sampling_strategy='all')

grid_search = GridSearchCV(estimator=brf, param_grid=param_grid, cv=10 ,scoring='f1')
grid_search.fit(X_resampled, y_resampled)
print(grid_search.best_params_)
```

# Entrenamiento del Modelo

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, stratify=y,random_state= seed)

smote_enn = SMOTEENN(random_state=seed)
X_resampled, y_resampled = smote_enn.fit_resample(X_train, y_train)

brf = BalancedRandomForestClassifier(n_estimators=280,max_depth=6,min_samples_split=2,random_state= seed,replacement=False,bootstrap=True,sampling_strategy='all')

# Entrenar el modelo
brf.fit(X_resampled, y_resampled)

y_scores = brf.predict_proba(X_test)[:, 1]

y_pred = (y_scores >= 0.5).astype(int)

# Evaluar el modelo
print(classification_report(y_test, y_pred))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['No Target', 'Target'], yticklabels=['No Target', 'Target'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

# Resultados

```python
# Calcular la curva ROC y el AUC
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
auc_score  = roc_auc_score(y_test, y_pred)

# Graficar la curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'AUC = {auc_score:.2f}')
plt.plot([0, 1], [0, 1], linestyle='--', color='red', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Curva ROC')
plt.legend()
plt.grid(True)
plt.show()
```

```python
y_train_pred_prob = brf.predict_proba(X_train)[:, 1]
y_test_pred_prob = brf.predict_proba(X_test)[:, 1]

auc_train = roc_auc_score(y_train, y_train_pred_prob)
print(f'AUC Train: {auc_train:.2f}')

# Calcular el AUC para el conjunto de prueba
auc_test = roc_auc_score(y_test, y_test_pred_prob)
print(f'AUC Test: {auc_test:.2f}')
```
